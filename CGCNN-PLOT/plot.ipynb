```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 14
plt.rcParams["axes.edgecolor"] = "black"
plt.rcParams["axes.linewidth"] = 4
df = pd.read_csv("CGCNN.csv")
df = df.iloc[:-1,:]

fig, ax = plt.subplots(figsize=(7, 4))
ax.errorbar(df.index+1, df['median'], 
            lolims= df['25Q'], uplims= df['75Q'],
            marker='o', markersize=8,
            linestyle='dotted')
ax.set_title("CGCNN Validation MAE\n(Median across 10 seeds)")
```
